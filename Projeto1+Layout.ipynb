```python
#Se usar o arquivo descompactado
#pd.read_csv('DOM2015.csv',sep=',')

#Se usar o arquivo compactado como está no dropbox púbico
#pd.read_csv('DOM2015csv.zip', compression='zip', header=0, sep=',', quotechar='"')
data = pd.read_csv('DOM2015.csv',sep=',')
```

```python
data.loc[(data.UF>44)&(data.UF<54),"REGIAO"]="Centro-Oeste"
#data.loc[(data.V4105>3),"URBANIZAÇÃO"]="Rural"

```

```python
data.REGIAO=data.REGIAO.astype('category')
data.V4622=data.V4622.astype('category')#renda\n"
data.V4622.cat.categories=('Sem Rendimento','até 0,25 SM','de 0,25 a 0,5 SM','de 0,5 a 1 SM','de 1 a 2 SM','de 2 a 3 SM','de 3 a 5 SM','mais que 5 SM','Não Disponível')
#data.V4105=data.V4105.astype('category')#urbano/rural\n",
data.V2322=data.V02322.astype('category')#acessoa a internet no domicilio\n",
data.V2322.cat.categories=('Sim','Não')
data.V02424=data.V02424.astype('category')#acesso via 3G e 4G\n",
data.V02424.cat.categories=('Sim','Não')
data.V02425=data.V02425.astype('category')#discada ou linha telefonica\n",
data.V02425.cat.categories=('Sim','Não')
data.V02426=data.V02426.astype('category')#acesso por banda larga\n",
data.V02426.cat.categories=('Sim','Não')
data.V0231=data.V0231.astype('category') #possui microcomputador
data.V0231.cat.categories=('Sim','Não') 
data.V0232=data.V0232.astype('category') #microcomputador para acessar a internet
data.V0232.cat.categories=('Sim','Não')
data = data[(data.REGIAO=="Centro-Oeste")]
```

```python
data=data.loc[:,["REGIAO","V4622","V02322","V02424","V02425","V02426","V0231","V0232"]]
#data.rename(columns = {"V4622":"RPC","V4105":"ZONA"})
data.columns=["REGIAO","RENDA","ACESSO À INTERNET PRÓPRIO DOMICILIO","VIA 3G OU 4G","DISCADA OU LINHA TELEFONICA","BANDA LARGA","MICROCOMPUTADOR","MICROCOMPUTADOR PARA INTERNET"]
data
```
