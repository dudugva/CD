## Gráficos:

A Internet Discada/Linha Telefônica pode ser considerada um dos mais primitivos meios de comunicação moderno que tivemos nos últimos anos. A conexão era baseada na linha telefônica e/ou em satélites para realizar a comunicação entre os computadores mesmo que em áreas urbanas. No nosso caso de análise da zona rural o caso fica ainda mais grave. Ao analisar a triplice de gráficos abaixo podemos ver a falta de uma linha telefonica sequer nas residências rurais do centro-oeste brasileiro. 

Além de já haver um número muito pequeno de pessoas que possuem linha telefonica ou internet discada, esse número foi reduzido ao longo do tempo, uma vez que outros meios de comunicação baseados na internet surgiram mais baratos e de mais simples instalação, visando este público alvo.

```python
#Grafico Discada 

fig = plt.figure(figsize=(15, 5))
plt.subplot(131)
plot = h2.plot(kind='pie',title='DISCADA OU LINHA TELEFONICA: 2013',autopct='%.1f')

plt.subplot(132)
plot = f2.plot(kind='pie',title='DISCADA OU LINHA TELEFONICA: 2014',autopct='%.1f')

plt.subplot(133)
plot = g2.plot(kind='pie',title='DISCADA OU LINHA TELEFONICA: 2015',autopct='%.1f')

plt.tight_layout()
plt.show()
```

Ao darmos um passo à frente, chegamos na conexão por banda larga. Este meio de comunicação está à frente da internet discada, mas não necessariamente atrás do 3G/4G uma vez que esta é realizada por cabos. A banda larga aparece no meio rural como primeiro meio de acesso à internet com maior estabilidade para o público e ao mesmo tempo não impossibilita o uso de telefone juntamente ao acesso à mesma. 

Podemos ver a crescente no uso de banda larga nos 3 anos abaixo, o que está fortemente ligado ao interesse das empresas de banda larga em investir na infraestrutura necessária para cabear longas extensões de terra, onde estão localizados os moradores rurais do centro-oeste.



```python
#Grafico Banda Larga

fig = plt.figure(figsize=(15, 5))
plt.subplot(131)
plot = h1.plot(kind='pie',title='BANDA LARGA: 2013',autopct='%.1f')

plt.subplot(132)
plot = f1.plot(kind='pie',title='BANDA LARGA: 2014',autopct='%.1f')

plt.subplot(133)
plot = g1.plot(kind='pie',title='BANDA LARGA: 2015',autopct='%.1f')

plt.tight_layout()
plt.show()
```

Chegamos enfim à infraesturura mais investida hoje no Brasil, a tecnologia mobile. O 3G e o 4G se mostraram extremamente importantes no cenário brasileiro por proporcionar à qualquer usuário pelo menos a comunicação por mensageiros instantâneos em tempo real. A cobertura já é realizada pelas empresas e quase 100% do território nacional tem cobertura de pelo menos uma das 4 maiores empresas de telefonia presentes no Brasil. 

O 3G/4G por sua vez tem um crescimento mais acelerado do que os vistos anteriormente. Na troca de um ano o número de usuários de 3G/4G subiu de 43% para 60.3%, um crescimento atronômico devido principalmente à busca de entregar planos de internet que condizam com o público rural brasileiro.

## Conclusão 

Por fim, é possível concluir que a conexão estabelecida pelo dial-up (discada) perde espaço devido ao seu preço e velocidade de acordo com a entrada dos tipos de conexão banda larga e rede móvel (3G/4G).

Essa dupla de conexões não se exclui, mas se complementa, uma vez que a rede móvel é, como próprio nome já diz, uma conexão contruída para funcionar nos celulares, e a banda larga, é uma internet com maior estabilidade, projetada para ser instalada em locais fixos. 

Com o desenvolvimento agropecuário da região do centro-oeste, consequentemente o desenvolvimento econômico acompanhou e consequentemente o desenvolvimento técnologico para conexões também. Nesse sentido, as empresas provedoras desses serviços puderam se voltar para esse público rural do centro-oeste e entregar planos com valores e benefícios que condizam com essa realizade e se tornem clientes das companhias.
